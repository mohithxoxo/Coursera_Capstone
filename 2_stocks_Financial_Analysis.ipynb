```
import yfinance as yf
tickerSymbol = 'TCS.NS'
tickerData = yf.Ticker(tickerSymbol)
TCS = tickerData.history(period='1d', start='2014-03-1', end='2020-03-31')
TCS
```

```
import yfinance as yf
tickerSymbol = 'INFY.NS'
tickerData = yf.Ticker(tickerSymbol)
INFY = tickerData.history(period='1d', start='2014-03-1', end='2020-03-31')
INFY
```

```
print(TCS.tail())
print(INFY.tail())
```

```
import matplotlib.pyplot as plt
%matplotlib inline
TCS['Close'].plot()
plt.show()
INFY['Close'].plot()
plt.show()
```

```
import pandas as pd
data = [TCS['Close'],INFY['Close']]
data = pd.DataFrame(data).transpose()
data = pd.DataFrame(data.values)
data = data.rename(columns={0:"TCS",1 : "INFY" })
data
```

```
data.plot(figsize=(10, 7))
plt.legend()
plt.title("Adjusted Close Price (31/03/2020) of TOP 2 IT companies ", fontsize=16)
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```

```
df = data 
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```

```
TCS['Volume'].plot(figsize=(16,8),label = "TCS",title='Volume')
INFY['Volume'].plot(label = "INFY")
plt.legend();
```

```
TCS['Total Traded'] = TCS['Open'] * TCS['Volume']
INFY['Total Traded'] = INFY['Open'] * INFY['Volume']
TCS['Total Traded'].plot(figsize=(16,8),label = "TCS",title='Total Money Traded')
INFY['Total Traded'].plot(label = "INFY")
plt.legend()
plt.ylabel('Total Money Traded')
```

```
TCS['returns'] = TCS['Close'].pct_change(1)
INFY['returns'] = INFY['Close'].pct_change(1)
```

```
TCS['returns'].hist(bins=100)
```

```
INFY['returns'].hist(bins=100)
```

```
TCS['returns'].hist(bins=50,label='TCS',figsize = (10,8), alpha= 0.5)
INFY['returns'].hist(bins=50,label='INFY', alpha= 0.5)
plt.xlabel('Returns')
plt.legend();
```

```
TCS['daily_cumulative_return'] = ( 1 + TCS['returns'] ).cumprod()
INFY['daily_cumulative_return'] = ( 1 + INFY['returns'] ).cumprod()
TCS['daily_cumulative_return'].plot(label='TCS',figsize = (16,8),title = 'Cumilative Return')
INFY['daily_cumulative_return'].plot(label='INFY')
plt.legend();
```

```
"""## TCS"""

print ("TCS returns mean = ", TCS['returns'].mean(),"\n\n",
       "TCS returns mean * 250 = ",TCS['returns'].mean()*250,"\n\n",
       "TCS returns Standard Deviation = ",TCS['returns'].std(),"\n\n",
       "TCS returns Standard Deviation * 250 * 0.5 = ",TCS['returns'].std()*250**0.5)
```

```
"""## Infosys"""

print ("Infosys returns mean = ", INFY['returns'].mean(),"\n\n",
       "Infosys returns mean * 250 = ",INFY['returns'].mean()*250,"\n\n",
       "Infosys returns Standard Deviation = ",INFY['returns'].std(),"\n\n",
       "Infosys returns Standard Deviation * 250 * 0.5 = ",INFY['returns'].std()*250**0.5)
```

```
"""## Final Results"""

print(TCS['returns'].mean()*250)
print(INFY['returns'].mean()*250)
```

```
TCS['returns'].plot(figsize=(10, 7))
INFY['returns'].plot(figsize=(10, 7))
plt.legend()
plt.title("Returns of Two companies ", fontsize=16)
plt.ylabel('Returns', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
