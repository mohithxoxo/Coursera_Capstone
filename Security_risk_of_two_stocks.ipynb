```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
import yfinance as yf
tickerSymbol = 'HDFCBANK.NS'
tickerData = yf.Ticker(tickerSymbol)
HDFCBANK = tickerData.history(period='1d', start='2010-03-1')
print("HDFCBANK shares " ,"\n",HDFCBANK.tail())

tickerSymbol = 'ICICIBANK.NS'
tickerData = yf.Ticker(tickerSymbol)
ICICIBANK= tickerData.history(period='1d', start='2010-03-1')
print("\n","ICICIBANK shares ","\n" ,ICICIBANK.tail())
```

```
HDFCBANK = HDFCBANK.rename(columns={'Close' : 'HDFC_Shares'})
ICICIBANK  = ICICIBANK.rename(columns={'Close' : 'ICICI_Shares'})
hdfc_icici_data = pd.DataFrame(data = [HDFCBANK['HDFC_Shares'],ICICIBANK['ICICI_Shares']]).T
hdfc_icici_data.plot(figsize=(10, 7))
plt.legend()
plt.title("Adjusted Close Price of Two companies ", fontsize=16)
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```

```
HDFCBANK_returns = np.log(HDFCBANK['HDFC_Shares'] / HDFCBANK['HDFC_Shares'].shift(1))
HDFCBANK_returns.tail(20)
```

```
ICICIBANK_returns = np.log(ICICIBANK['ICICI_Shares'] / ICICIBANK['ICICI_Shares'].shift(1))
ICICIBANK_returns.tail(20)
```

```
"""## HDFC"""

print ("HDFC returns mean = ", HDFCBANK_returns.mean(),"\n\n",
       "HDFC returns mean * 250 = ",HDFCBANK_returns.mean()*250,"\n\n",
       "HDFC returns Standard Deviation = ",HDFCBANK_returns.std(),"\n\n",
       "HDFC returns Standard Deviation * 250 * 0.5 = ",HDFCBANK_returns.std()*250**0.5)
```

```
"""## ICICI"""

print ("ICICI returns mean = ", ICICIBANK_returns.mean(),"\n\n",
       "ICICI returns mean * 250 = ",ICICIBANK_returns.mean()*250,"\n\n",
       "ICICI returns Standard Deviation = ",ICICIBANK_returns.std(),"\n\n",
       "ICICI returns Standard Deviation * 250 * 0.5 = ",ICICIBANK_returns.std()*250**0.5)
```

```
"""## Final Results"""

print(HDFCBANK_returns.mean()*250)
print(ICICIBANK_returns.mean()*250)
```

```
hdfc_icici_returns = pd.DataFrame(data = [HDFCBANK_returns,ICICIBANK_returns]).T
print ("hdfc vs icici mean * 250","\n\n",hdfc_icici_returns[['HDFC_Shares','ICICI_Shares']].mean()*250,"\n\n\n\n",
"hdfc vs icici mean * 250 square 0.5 ","\n\n",hdfc_icici_returns[['HDFC_Shares','ICICI_Shares']].mean()*250**0.5)
```

```
hdfc_icici_returns.plot(figsize=(10, 7))
plt.legend()
plt.title("Returns of Two companies ", fontsize=16)
plt.ylabel('Returns', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
